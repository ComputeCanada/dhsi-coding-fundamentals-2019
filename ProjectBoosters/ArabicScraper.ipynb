First thing: can we use the requests library to pull content from the page

```python
import requests

r = requests.get('https://ar.wikisource.org/wiki/%D9%85%D8%AE%D8%AA%D8%A7%D8%B1_%D8%A7%D9%84%D8%B5%D8%AD%D8%A7%D8%AD/%D9%81%D9%87%D8%B1%D8%B3')
r.status_code

#r.headers['content-type']

#r.encoding

#r.text

r.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')

soup.find_all('h2')
#len(soup.find_all('h2'))
```

I know you want the H2 header content and we can see that there are 29 such headers in the document but I'm not sure what you want besides this.  Here's how you get the text:

```python
for h2 in soup.find_all('h2'):
    print(h2.get_text())
```

Here's how to get the IDs:

```python
for h2 in soup.find_all("span", class_="mw-headline"):
    print(h2.get('id'))
```
